참고를 위해 Qiime2 코드를 업로드:

1.	Data import

conda activate qiime2-2021.8

qiime tools import \
  --type 'SampleData[PairedEndSequencesWithQuality]' \
  --input-path ./fastq \
  --input-format CasavaOneEightSingleLanePerSampleDirFmt \
  --output-path final-casavapaired.qza

2.	Visualization of sequencing quality
qiime demux summarize \
--i-data ./final-casavapaired.qza \
--o-visualization ./final-casavapaired.qzv

3.	Denoising with DADA2
 Visualization of sequencing quality 보고 trim, trunc는 판단. Quality score 20~25로 진행.

qiime dada2 denoise-paired \
--i-demultiplexed-seqs final-casavapaired.qza \
--p-trim-left-f 10 \
--p-trim-left-r 10 \
--p-trunc-len-f 220 \
--p-trunc-len-r 180 \
--o-table table_final5.qza \
--o-representative-sequences rep-seqs_final-casavapaired5.qza \
--o-denoising-stats denoising-stats_final-casavapaired5.qza

4.	Viualization of DADA2 denosing result
qiime metadata tabulate \
--m-input-file ./denoising-stats_final-casavapaired5.qza \
--o-visualization ./denoising-stats_final-casavapaired5.qzv

qiime feature-table tabulate-seqs \
--i-data rep-seqs_final-casavapaired5.qza \
--o-visualization rep-seqs_final-casavapaired5.qzv


위 과정이 끝난 후 https://view.qiime2.org/ 에서 rep-seqs_final-casavapaired5.qzv 파일을 열어 denoising 과정까지 끝난 raw fasta file을 다운로드. (= sample.fasta)

```python
import pandas as pd
depileup = pd.read_csv('DE_sorted.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
depileup.tail()
```

```python
import re
toremove = re.compile('[<>$*#^]')
depileup['matches'] = depileup['basereads'].apply(lambda x: toremove.sub('', x))
```

```python
depileup[['chrom', 'pos', 'matches']]
```

```python
depileup2 = depileup[['chrom', 'pos', 'matches']]
```

```python
depileup3 = depileup2['matches']
```

```python
depileup2['base'] = depileup3.str.len()
```

```python
depileup2
```

```python
import math
from collections import Counter

def entropy(s):
	p, lns = Counter(s), float(len(s))
	return -sum( count/lns * math.log(count/lns, 2) for count in p.values())
```

```python
count = ['1','2']
dent = pd.Series(count)
print(dent)
print(dent[1])
```

```python
i = 0
while i <= 2828:
  dent[i] = entropy(depileup3[i])
  i += 1
```

```python
depileup4 = depileup2
depileup4['Shannon_entropy'] = dent
print(depileup4)
```

```python
depileup5 = depileup4[['chrom','pos']]
depileup5['pos2']=depileup4['pos']+1
depileup5['Shannon_entropy'] = depileup4['Shannon_entropy']
```
